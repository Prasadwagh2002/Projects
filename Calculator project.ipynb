```python
import tkinter as tk
```

```python
# Backend part
calculation =""
# clicking the buttons
def click (symbol):
    global calculation
    calculation+= str (symbol)
    display.delete(1.0,"end")
    display.insert(1.0,calculation)
# calculation of result
def result():
    try:
        global calculation
        calculation = str(eval(calculation))
        display.delete(1.0,"end")
        display.insert(1.0,calculation)
    except:
        # to handle the error part
        clear()
        display.insert(1.0,"error")
# clear the display
def clear():
    global calculation 
    calculation = " "
    display.delete(1.0,"end")
# Backspace the display
def Backspace(length):
    global calculation
    calculation=calculation [:-1]
    display.delete(1.0,"end")
    display.insert(1.0,calculation)
    
```

```python
# initiating the application
app = tk.Tk()

# shape (height and width ) of window
app.geometry("497x310")

# title of application
app.title("Prasad Calculator")


# display
display = tk.Text(app,height = 3, width = 15, bg = "white",font = ("Arial",24,"bold"))
display.grid(row=1,columnspan=4)

# buttons 0-9
button_1 = tk.Button(app,text="1",width=4,fg="orange",bg="black",font=("Arial,14"),command = lambda:click(1))
button_1.grid(row=4,column=0,padx=1,pady=2)

button_2 = tk.Button(app,text="2",width=4,fg="orange",bg="black",font=("Arial,14"),command = lambda:click(2))
button_2.grid(row=4,column=1,padx=1,pady=2)

button_3 = tk.Button(app,text="3",width=4,fg="orange",bg="black",font=("Arial,14"),command = lambda:click(3))
button_3.grid(row=4,column=2,padx=1,pady=2)


button_4 = tk.Button(app,text="4",width=4,fg="orange",bg="black",font=("Arial,14"),command = lambda:click(4))
button_4.grid(row=3,column=0,padx=1,pady=2)

button_5 = tk.Button(app , text = "5", width = 4,fg= "orange",bg ="black",font = ("Arial,14"),command = lambda:click(5))
button_5.grid(row = 3, column = 1, padx =1, pady =2)

button_6 = tk.Button(app , text = "6", width = 4,fg= "orange",bg ="black",font = ("Arial,14"),command = lambda:click(6))
button_6.grid(row = 3, column = 2, padx =1, pady =2)


button_7 = tk.Button(app , text = "7", width = 4,fg = "orange",bg ="black",font = ("Arial,14"),command = lambda:click(7))
button_7.grid(row = 2,column = 0,padx =1, pady =2)

button_8 = tk.Button(app , text = "8", width = 4,fg= "orange",bg ="black",font = ("Arial,14"),command = lambda:click(8))
button_8.grid(row = 2,column = 1,padx =1, pady =2)

button_9 = tk.Button(app , text = "9", width = 4,fg= "orange",bg ="black",font = ("Arial,14"),command = lambda:click(9))
button_9.grid(row = 2,column = 2,padx =1, pady =2)

button_0 = tk.Button(app , text = "0", width = 4,fg= "orange",bg ="black",font = ("Arial,14"),command = lambda:click(0))
button_0.grid(row = 5, column = 1,padx =1, pady =2)

# Symbolic buttons

button_plus = tk.Button(app , text = "+", width = 4,fg= "orange",bg ="black",font = ("Arial,14"),command = lambda:click("+"))
button_plus.grid(row = 2, column = 3,padx =1, pady =2)

button_minus = tk.Button(app , text = "-", width = 4,fg= "orange",bg ="black",font = ("Arial,14"),command = lambda:click("-"))
button_minus.grid(row = 3, column = 3,padx =1, pady =2)

button_multiply = tk.Button(app , text = "*", width = 4,fg= "orange",bg ="black",font = ("Arial,14"),command = lambda:click("*"))
button_multiply.grid(row = 4, column = 3,padx =1, pady =2)

button_divide = tk.Button(app , text = "/", width = 4,fg= "orange",bg ="black",font = ("Arial,14"),command = lambda:click("/"))
button_divide.grid(row = 5, column = 3,padx =1, pady =2)

button_clear = tk.Button(app , text = "c", width = 4,fg= "orange",bg ="black",font = ("Arial,14"),command = clear)
button_clear.grid(row = 5, column = 0,padx =1, pady =2)

button_equal = tk.Button(app , text = "=", width = 4,bg="blue",font = ("Arial,14"),command = result)
button_equal.grid(row = 5, column = 2,padx =1, pady =2)

button_Backspace = tk.Button(app , text = "<-", width = 4,bg="blue",font = ("Arial,14"),command = lambda:Backspace("<-"))
button_Backspace.grid(row = 6, column = 3,padx =1, pady =2)


app.mainloop()
```
